```python
Vip = orders[orders['amount_spent']>orders['amount_spent'].quantile(0.95)]
Vip
```

## Q2: How to identify which country has the most VIP Customers?

```python
#Sub Problem 1: How to aggregate the  amount_spent for unique customers?
orders.groupby('CustomerID')['amount_spent'].agg(['count','sum','max','min','std','mean'])
orders
```

```python
#Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?
percentil = .75


customer = orders[orders['amount_spent']>orders['amount_spent'].quantile(percentil)]


customer.groupby('CustomerID')['amount_spent'].agg(['count','sum','max','min','std','mean'])
```
